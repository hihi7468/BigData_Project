```python
import urllib.request
import bs4
import requests
base_url = "https://finance.naver.com/research/company_list.naver"
response = requests.get(base_url)
html = response.text
soup =  bs4.BeautifulSoup(html, 'html.parser') ## 설명 생략
tr_set = soup.findAll("tr")
th_box = tr_set[0].findAll("th")
top_name = []
for i in range(0, len(th_box)):
    top_name.append(th_box[i].text)
#print(c) #3번쨰부터 10번째,
for i in range(0, len(top_name)):
    print(top_name[i], end= " ")
print("\n")
count = 0
for i in range(2, (len(tr_set)-5)):
    count += 1 
    if count == 8:
        count = 0
    elif count != 6 and count != 7 and count != 8:
        d = tr_set[i].findAll("a")
        e = tr_set[i].findAll("td")
        #print(a[i]) #a[i].text
        print(d[0].text ,d[1].text, e[2].text, e[3].img["alt"], e[4].text, e[5].text )
```
