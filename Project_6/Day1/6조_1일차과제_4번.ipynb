필요한 <td>태그 안의 값들이 찍히는것을 확인 할 수 있다. 해당 값들을 리스트로 담자.

```python
name_value = []
for i in range(2, len(money_name)):
    name_value = [] # 값 초기화
    for j in money_name[i].findAll("td"):
        name_value.append(j.text.replace("\n", ""))
    print(name_value)
```

\n 말고 \t의 값들도 보인다 .\t의 값들도 str 에서 미리 공백으로 대체하자.

```python
name_value = []
for i in range(2, len(money_name)):
    name_value = [] # 값 초기화
    for j in money_name[i].findAll("td"):
        name_value.append(j.text.replace("\t", "").replace("\n", ""))
    print(name_value)
```

이제 모든 코드를 합치면 아래와 같이 된다.

```python
import urllib.request
import bs4
import requests
base_url = "https://finance.naver.com/marketindex/exchangeList.naver"
response = requests.get(base_url)
html = response.text
soup =  bs4.BeautifulSoup(html, 'html.parser')
money_name = soup.findAll("tr")
money_name_list = []
for i in money_name[0].findAll("th"):
    money_name_list.append(i.text.replace("\n", "")) # 통화면의 경우 \n통화명으로 담기어서 str에서 \n값을 공백으로 대체해서 주기로 함.
for i in money_name[1].findAll("th"):
    money_name_list.append(i.text.replace("\n", ""))
print(money_name_list[0]+"     "+ money_name_list[1]+"     "+money_name_list[2]+"        "+money_name_list[3]+"              "+money_name_list[4])
print("                            "+money_name_list[5]+money_name_list[6]+money_name_list[7]+money_name_list[8])
name_value = []
for i in range(2, len(money_name)):
    name_value = [] # 값 초기화
    for j in money_name[i].findAll("td"):
        name_value.append(j.text.replace("\t", "").replace("\n", ""))
    print(name_value)
```
