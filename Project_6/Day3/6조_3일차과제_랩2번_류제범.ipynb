```python
import openpyxl
from selenium import webdriver
import time
import pandas as pd
from urllib.request import urlopen
import json
url = "http://www.krei.re.kr:18181/chart/main_chart/fob/sdate/2000-01-01/edate/2022-02-07"
data = urlopen(url)
data.read()
html = urlopen(url)
json_obj = json.load(html) 
raw_data=[]
print(json_obj)
for item in json_obj:
    raw_data.append([item['id'],item['usa_fob'],item['thailand_fob'],item['date']])
raw_data=pd.DataFrame(raw_data)
print(raw_data)
raw_data.to_excel(excel_writer='tt.xlsx')
```
