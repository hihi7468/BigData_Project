```python
driver = webdriver.Chrome("chromedriver.exe")#r"C:\Users\82104\chromedriver/chromedriver.exe"
driver.get("https://www.billboard.com/charts/hot-100/")
```

```python
elem = driver.find_elements(By.CLASS_NAME,'chart-results')
print(len(elem)) # elem은 find.elements를 통하여 findall과 같다. 길이를 통하여 해당 chart-results를 가지고 있는게, 몇개인지 먼저 찍어보자.
#해당 결과를 통하여 chart-result값을 가지는 div class 속성은 한개 밖에 없음을  알 수 있다.

a = elem[0].find_elements(By.CLASS_NAME,'u-max-width-960')
print(len(a))
b = a[0].find_elements(By.CLASS_NAME, "lrv-u-flex")
print(len(b))
print(b[0].text)
print("\n")
c= b[0].find_elements(By.CLASS_NAME, "c-tagline")
print(len(c))
print(c[0].text)
```

```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import psycopg2
conn_str = "host=localhost dbname=postgres user=postgres password=admin port=5432"
conn = psycopg2.connect(conn_str)
driver = webdriver.Chrome("chromedriver.exe")
base_url = "https://www.billboard.com/charts/hot-100/"
set_year = int(input("선택할 년도와 월 그리고 일을 넣어주세요. ex)20210101\n"))
end_year = int(input("끝낼 년도와 월 그리고 일을 넣어주세요. ex)20210101\n"))
month_length = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
year= 2022
get_month=int(str(set_year)[4:6])
get_day =int(str(set_year)[6:8])
get_end_month=int(str(end_year)[4:6])
get_end_day=int(str(end_year)[6:8])
total_length= 0
cur = conn.cursor()
cur.execute('''CREATE TABLE HOT100 (number serial, wk varchar(255), rank varchar(255), title varchar(255), singer varchar(255), date varchar(255))''') 
conn.commit()
for k in range(get_month, get_end_month+1):
    if k == get_month:
        total_length += (month_length[get_month-1] - get_day+1)
    elif k !=get_month and k != get_end_month:
        total_length +=(month_length[k-1])
    elif k == get_end_month:
        total_length += (get_end_day)

focus_cursor_month= get_month
focus_cursor_day=get_day
for x in range(total_length):
    if len(str(focus_cursor_day)) == 1:
        day="0"+str(focus_cursor_day)
    elif len(str(focus_cursor_day)) == 2:
        day = str(focus_cursor_day)
        
    if len(str(focus_cursor_month)) == 1:
        month="0"+str(focus_cursor_month)
    elif len(str(focus_cursor_month)) == 2:
        month= str(focus_cursor_month)
        
    if month_length[focus_cursor_month-1] != focus_cursor_day:
        print(focus_cursor_day)
        focus_cursor_day += 1
    elif month_length[focus_cursor_month-1] == focus_cursor_day:
        focus_cursor_day=1
        focus_cursor_month +=1
        
    url = base_url+str(year)+"-"+month+"-"+day+"/"
    driver.get(url)
    elem = driver.find_elements(By.CLASS_NAME, "chart-results")[0].find_elements(By.CLASS_NAME,'u-max-width-960')[0].find_elements(By.CLASS_NAME, "lrv-u-flex")[0].find_elements(By.CLASS_NAME, "c-tagline")
    set_root= driver.find_elements(By.CLASS_NAME, "o-chart-results-list-row-container")
    for i in range(0, (len(set_root))):
        get_rank = set_root[i].find_elements(By.CLASS_NAME, "o-chart-results-list-row")[0].find_elements(By.CLASS_NAME, "o-chart-results-list__item")[0].find_elements(By.CLASS_NAME, "c-label")[0].text
        get_title_name = set_root[i].find_elements(By.CLASS_NAME, "o-chart-results-list-row")[0].find_elements(By.CLASS_NAME, "o-chart-results-list__item")[3].find_elements(By.ID,"title-of-a-story")[0].text
        get_title_singer = set_root[i].find_elements(By.CLASS_NAME, "o-chart-results-list-row")[0].find_elements(By.CLASS_NAME, "o-chart-results-list__item")[3].find_elements(By.CLASS_NAME,"c-label")[0].text
        date = str(year)+"-"+month+"-"+day
        cur.execute("INSERT INTO HOT100 (wk, rank, title, singer, date) VALUES (%s, %s, %s, %s, %s);",(elem[0].text, get_rank, get_title_name, get_title_singer, date))
        conn.commit()
        print(get_rank)
        print(get_title_name)
        print(get_title_singer)
        print(elem[0].text)
        print("\n")
```
