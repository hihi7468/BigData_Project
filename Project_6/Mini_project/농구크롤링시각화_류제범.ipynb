```python
test = df1.copy()
test2 = df2.copy()
team_away_list =list(test2['Team_name1'].unique()) # 어웨이 팀에서의 출전팀과 홈팀에서의 출전팀의 경우 경기가 작은 팀들은
team_home_list =list(test2['Team_name2'].unique()) #어웨이팀에서는 없을 수도 있음. 따라서 두개를 비교해보고자 함.
team_all_list= [] 
print(team_away_list) #어웨이 팀으로 출전한 팀 명단을 봅니다.
print(team_home_list) #홈 팀으로 출전한 팀 명단을 봅니다.
print(len(team_away_list)) #어웨이 팀으로 출전한 팀의 개수
print(len(team_home_list)) #홈팀으로 출전한 팀의 개수
for i in range(0, len(team_away_list)):
    team_all_list.append(team_away_list[i]) # 일단은 어웨팀으로 출전한 팀 전체 이름을 넣습니다.
for i in range(0, len(team_home_list)):
    team_all_list.append(team_home_list[i])   #어웨이팀, 홈팀 둘다에 속하여 중첩되는 팀 이름도 있을 것입니다.
print(team_all_list) # 중첩된 팀 리스트 출력
print(len(team_all_list)) #42+35 = 77
team_all_list= set(team_all_list) #중첩된 팀들을 제거합니다.
print(team_all_list) # 중첩된것을 제거하고 47개의 팀들이 있음을 알 수 있습니다. 그리하여, 77개중 중첩된 30개가 빠졋습니다.
print(len(team_all_list))
```

```python
team_all_list= sorted(team_all_list)  #팀들의 이름을 A~Z 순으로 정렬합니다.
print(team_all_list) # 정렬된 팀들과 팀들의 개수를 확인합니다.
print(len(team_all_list))
```

```python
team_all_game_count= [] #팀 명단에 맞춰서 이제 각 팀들이 몇번 뛰었는지 체크하기 위함.어웨이팀에서의 카운트+홈팀에서의 카운트
for i in range(len(team_all_list)): #팀 이름과 경기 횟수를 매칭 시키기 위해서 team_all_list의 개수만큼 game_count도 0으로 셋팅
    team_all_game_count.append(0)
for i in range(0, len(team_all_list)): # 모든 팀의 이름을 가지고 있는 리스트의 크기만큼 포문을 돌립니다.
    for j in range(0, len(team_away_list)): # 만약 어웨이팀에 출전하지 않은 경우에 이름을 넣는다면 없는 key이기때문에 에러가 발생.
        if team_all_list[i] == team_away_list[j]:  #따라서 if로 해당 팀이름이 어웨이 team의 모든 이름을 찾아서 
            print(test2['Team_name1'].value_counts()[team_all_list[i]]) #모든 팀명단에 있는 이름이 어웨이 팀명단에도 있다면 넣어줍니다.
            team_all_game_count[i] += test2['Team_name1'].value_counts()[team_all_list[i]]
print("자 이제, away 팀의 기록을 체크해봅시다.")
print(team_all_game_count)
```

자 이제, 홈팀과 어웨이 팀의 승률을 분석하기 위하여
홈팀에서의 경기 판수 , 홈팀에서 이긴 판수 // 어웨이팀에서의 경기 판수, 어웨이팀에서 이긴 판수 를 분석하겠다.
team_win_count의 변수가 아닌 새롭게 team_win_home_count와 team_win_away_count를 만들겠다.
또한, 홈, 어웨이 경기횟수를 담는것을 만들어야 한다.

```python
team_win_home_count=[]
team_win_away_count=[]
team_home_game_count=[]
team_away_game_count=[]
for i in range(0, len(team_all_list)):
    team_win_set1=test2[test2['Team_name1']==team_all_list[i]]['team1_win'].value_counts()[1]
    team_win_set1_sub=test2[test2['Team_name1']==team_all_list[i]]['team1_win'].value_counts()[0]
    team_win_away_count.append(team_win_set1)
    team_away_game_count.append(team_win_set1+team_win_set1_sub)
    
    team_win_set1=test2[test2['Team_name2']==team_all_list[i]]['team2_win'].value_counts()[1]
    team_win_set1_sub=test2[test2['Team_name2']==team_all_list[i]]['team2_win'].value_counts()[0]
    team_win_home_count.append(team_win_set1)
    team_home_game_count.append(team_win_set1+team_win_set1_sub)
print(team_win_home_count)
print(team_win_away_count)
print(team_home_game_count)
print(team_away_game_count)
```

```python
team_win_home_rate=[]
team_win_away_rate=[]
for i in range(0, len(team_all_list)):
    team_win_home_rate.append((team_win_home_count[i]/team_home_game_count[i])*100)
    team_win_away_rate.append((team_win_away_count[i]/team_away_game_count[i])*100)
print(team_win_home_rate)
print(team_win_away_rate)
```

```python
plt.figure(figsize=(30,10), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title('각 팀별 home 승률 rate')
plt.bar(team_all_list, team_win_home_rate)
plt.plot(team_all_list, low_team_line, color='red')
plt.plot(team_all_list, normal_team_line, color='black')
plt.plot(team_all_list, high_team_line, color='green')
plt.show()
```

```python
plt.figure(figsize=(30,10), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title('각 팀별 away 승률 rate')
plt.bar(team_all_list, team_win_away_rate)
plt.plot(team_all_list, low_team_line, color='red')
plt.plot(team_all_list, normal_team_line, color='black')
plt.plot(team_all_list, high_team_line, color='green')
plt.show()
```
