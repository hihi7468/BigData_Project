```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

populations = pd.DataFrame({'0-9age':[80, 120], '10-19age':[100,200], '20-29age':[140,300], '30-39age':[160,360],
                                  '40-49age':[200,400], '50-59age':[240, 300], '60-69age':[195, 40], '70-79age':[160, 25],
                                  '80-89age':[80, 20], '90-99age':[20, 10]},
                           index = ['CityA', 'CityB'],
                           columns = ['0-9age', '10-19age', '20-29age', '30-39age','40-49age', '50-59age', '60-69age', '70-79age', '80-89age', '90-99age']
                           ) 
citya_keylist = populations.keys()
citya_valuelist = populations.loc['CityA']
cityb_valuelist = populations.loc['CityB']
```

```python
def draw_barchart(citya_keylist, citya_valuelist, cityb_valuelist):
    x_labels = citya_keylist

    index = np.arange(len(x_labels))
    width = 0.4

    fig, ax = plt.subplots()
    rect1 = ax.bar(index - width/2, citya_valuelist, width, label="City A")
    rect2 = ax.bar(index + width/2, cityb_valuelist, width, label="City B")

    ax.set_xlabel('age')
    ax.set_ylabel('Population')                     # y축 레이블
    ax.set_title("The demographics of two cities")  # 그래프 타이틀

    plt.xticks(index + (width/10), x_labels)                   # X축 데이터
    #plt.legend(fontsize=9, loc="upper right")
    #plt.savefig('barchart01.png')
    #plt.close()
    fig
draw_barchart(citya_keylist, citya_valuelist, cityb_valuelist)
```

```python
citya_total = 0
citya_value = 0
cityb_value = 0
cityb_total = 0
for i in range(0, len(citya_valuelist)):
    citya_total += populations.loc['CityA'].values[i]
for i in range(0, len(citya_valuelist[populations.loc['CityA'].keys()> '10-19age'])):
    citya_value += citya_valuelist[populations.loc['CityA'].keys() > '10-19age'].values[i]
for i in range(0, len(cityb_valuelist[populations.loc['CityB'].keys()> '10-19age'])):
    cityb_value += cityb_valuelist[populations.loc['CityB'].keys() > '10-19age'].values[i]
for i in range(0, len(citya_valuelist)):
    cityb_total += populations.loc['CityB'].values[i]
title_name = "Voting rate"
print(citya_value)
print(cityb_value)
print(cityb_total)
print(len(cityb_valuelist[populations.loc['CityB'].keys()> '10-19age']))
```

```python
def draw_piechart(citya_total, citya_value, cityb_total, cityb_value, title_name):
    '''
        Pic chart를 그려줌
        citya_total: City A's total population
        citya_value: City A's a specific value

        cityb_total: City B's total population
        cityb_value: City B's a specific value
    '''
    citya_list = [citya_value, citya_total-citya_value]
    cityb_list = [cityb_value, cityb_total-cityb_value]
    label_list = [title_name , ""]

    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # City A의 pie chart
    ax1.pie(citya_list, labels=label_list, autopct="%.1f%%")
    ax1.set_title("City A's " +  title_name)

    # City B의 pie chart
    ax2.pie(cityb_list, labels=label_list, autopct="%.1f%%")
    ax2.set_title("City B's " + title_name)
    plt.savefig(title_name+".png")
    fig
    #plt.close()
draw_piechart(citya_total, citya_value, cityb_total, cityb_value, title_name)
```

```python
citya_value2 = 0
cityb_value2 = 0
for i in range(0, len(citya_valuelist[populations.loc['CityA'].keys()> '60-69age'])):
    citya_value2 += citya_valuelist[populations.loc['CityA'].keys() > '60-69age'].values[i]
for i in range(0,len(cityb_valuelist[populations.loc['CityB'].keys()> '60-69age'])):
    cityb_value2 += cityb_valuelist[populations.loc['CityB'].keys() > '60-69age'].values[i]
title_name = "Old people rate"
```

```python
def draw_piechart(citya_total, citya_value2, cityb_total, cityb_value2, title_name):
    '''
        Pic chart를 그려줌
        citya_total: City A's total population
        citya_value: City A's a specific value

        cityb_total: City B's total population
        cityb_value: City B's a specific value
    '''
    citya_list = [citya_value2, citya_total-citya_value2]
    cityb_list = [cityb_value2, cityb_total-cityb_value2]
    label_list = [title_name , ""]

    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

    # City A의 pie chart
    ax1.pie(citya_list, labels=label_list, autopct="%.1f%%")
    ax1.set_title("City A's " +  title_name)

    # City B의 pie chart
    ax2.pie(cityb_list, labels=label_list, autopct="%.1f%%")
    ax2.set_title("City B's " + title_name)
    plt.savefig(title_name+".png")
    fig
    #plt.close()
draw_piechart(citya_total, citya_value2, cityb_total, cityb_value2, title_name)
```
